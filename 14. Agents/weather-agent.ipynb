```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
from langchain_core.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun
import requests
```

```python
search_tool = DuckDuckGoSearchRun()
```

```python
@tool
def weather_api(city: str) -> float :
    """
    Given the city name the function return the temperature of the city.
    """
    url = f'http://api.weatherstack.com/current?access_key=ad61e6dc872afe58e5fe1778996548dd&query={city}'

    response = requests.get(url)

    return response.json()
```

```python
# weather_api.invoke('Lucknow')
```

```python
llm = ChatGoogleGenerativeAI(model='gemini-2.5-flash')
```

```python
agent = create_agent(
    model=llm,
    tools=[weather_api, search_tool]
)
```

```python
result = agent.invoke({
    'messages': [{'role': 'user', 'content': 'What is capital city of Uttar Pradesh also tell me about its current weather(temperature, humidity) ?'}]
})
```
